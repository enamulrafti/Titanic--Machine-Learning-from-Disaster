One hot encoder in Sex attribute

```python
one_hot_en_sex=pd.get_dummies(train_data['Sex'],drop_first=True)
```

```python
feature_scale = ['Age','Fare']
train_data[feature_scale] = scl.fit_transform(train_data[feature_scale])
```

```python
x = train_data.drop(['Survived'],axis=1)
y = train_data['Survived']
```
